## Reused functions from Assignment 1

```python
# Copy and paste functions from Assignment 1 here that you need for this assignment

#self.column_filter   - a column filter (see Assignment 1) from df

def create_column_filter(df):
    # use dropna, how='all' remove the columns with all NaN values
    filtered_df = df.dropna(axis=1, how='all')
    
    # use apply to cal the unique values in columns and find columns where this value==1
    nunique = df.apply(pd.Series.nunique)
    same_value_cols = nunique[nunique == 1].index
    filtered_df = filtered_df.drop(same_value_cols, axis=1)
    
    column_filter = filtered_df.columns.tolist()
    return filtered_df, column_filter

def apply_column_filter(df, column_filter):
    return df[column_filter]

# self.imputation      - an imputation mapping (see Assignment 1) from df

def create_imputation(df):
    imputation = {}
    cols = [col for col in df.columns if col != "CLASS" and col != "ID"]
    for col in cols:
        # numeric types: int64 and float64
        if df[col].dtype == 'int64' or df[col].dtype == 'float64':
            # all nan value
            if df[col].isnull().all():
                avg = 0

            else:
                avg = df[col].mean()
            imputation[col] = avg

        # other types
        else:
            imputation[col] = df[col].mode().iloc[0]
    df = df.fillna(value=imputation)
    return df, imputation

def apply_imputation(df, imputation):
    df = df.fillna(value=imputation)
    return df

# self.normalization   - a normalization mapping (see Assignment 1), using normalizationtype from the imputed df

def create_normalization(df,normalizationtype='minmax'):
    new_df=df.copy()
    cols=new_df.columns.tolist()
    cols=[col for col in cols if col!="CLASS" and col!="ID"]
    if normalizationtype=='minmax':
        mins=new_df[cols].min()
        maxs=new_df[cols].max()
#         print(mins,maxs)
        types=[normalizationtype]*len(cols)
        minmax=zip(types,mins,maxs)
        normalization=dict(zip(cols,minmax))
        new_df[cols]=(new_df[cols]-mins)/(maxs-mins)
    
    if normalizationtype=='zscore':
        means=new_df[cols].mean()
        stds=new_df[cols].std()
        #print(mins,maxs)
        types=[normalizationtype]*len(cols)
        zscore=zip(types,means,stds)
        normalization=dict(zip(cols,zscore))
        new_df[cols]=(new_df[cols]-means)/(stds)   
    
    return new_df,normalization

def apply_normalization(df,norm):
    new_df=df.copy()
    cols=new_df.columns.tolist()
    cols=[col for col in cols if col!="CLASS" and col!="ID"]
    for col in cols:
        new_df[col]=(new_df[col]-norm[col][1])/(norm[col][2]-norm[col][1])
    return new_df

# self.one_hot         - a one-hot mapping (see Assignment 1)

from pandas.api.types import is_numeric_dtype
def create_one_hot(df):
    df = df.copy()
    one_hot = {}
    for col in df.columns:
        if col != 'ID' and col != 'CLASS':
            if not is_numeric_dtype(df[col]):
                df[col] = df[col].astype("category")
                one_hot[col] = list(df[col].cat.categories)
                for i in one_hot[col]:
                    new_col = df[col]==i
                    new_col = new_col.astype("float")
                    df[col + "_" + i] = new_col 
                df = df.drop(columns = col, axis = 1)
    return df, one_hot

def apply_one_hot(df,one_hot):
    df = df.copy()
    for col in df.columns:
        if col in one_hot.keys():    
            for i in one_hot[col]:
                new_col = df[col]==i
                new_col = pd.Series(new_col.astype("float"))
                df[col + "_" + i] = new_col
            df = df.drop(columns = col, axis = 1)          
    return df  

def brier_score(df,correctlabels):
    new_df=df.copy()
    score=0.0
    
    for index, row in new_df.iterrows():
        c_index=np.where(new_df.columns==correctlabels[index])[0]
        #print(c_index,correctlabels[index],row)
        row.iloc[c_index]-=1
        
    new_df=new_df**2
    score=new_df.sum()
    score=score.sum()/len(correctlabels)
    return score

def accuracy(df, correctlabels):
    pred = df.idxmax(axis=1)
    accuracy = len(np.where(correctlabels==pred)[0]) / len(correctlabels)
    return accuracy

def find_pos_neg_instance_scores(prediction_column, correct_labels, column_name):
    prediction_column_array = np.array(prediction_column)
    binary_AUC = [column_name == i for i in correct_labels] 
    #print('binary_AUC=', binary_AUC)
    # positive_instance_scores => Col-Name (predicted class) matches with the correct class
    positive_instance_scores = prediction_column_array[binary_AUC]
    # negative_instance_scores => Col-Name (predicted class) DOES NOT match with the correct class
    negative_instance_scores = prediction_column_array[~np.array(binary_AUC)]
    return positive_instance_scores, negative_instance_scores

# fpr=False_Positive_Rate
# tpr=True_Positive_Rate
def get_tpr_fpr(prediction_column, correct_labels, column_name):
    positive_instance_scores, negative_instance_scores = find_pos_neg_instance_scores(prediction_column, correct_labels, column_name)
#     print("positive_instance_scores=",positive_instance_scores)
#     print("negative_instance_scores=",negative_instance_scores)
    scores_extended = [0]
    scores_extended += sorted(prediction_column)
    scores_extended += [1]
#     print('scores_extended=',scores_extended)
    dict_tpr_fpr = {}
    for predicted_score in scores_extended:
        if(len(negative_instance_scores)==0):
            count_false_pos_instances = 0
        else:
            count_false_pos_instances = np.sum(negative_instance_scores >= predicted_score)/len(negative_instance_scores)
        
        if(len(positive_instance_scores)==0):
            count_true_pos_instances = 0
        else:
            count_true_pos_instances = np.sum(positive_instance_scores >= predicted_score)/len(positive_instance_scores)
            
        #dict_tpr_fpr[predicted_score] = [count_false_pos_instances, count_true_pos_instances]
        dict_tpr_fpr[predicted_score] = [count_true_pos_instances, count_false_pos_instances]
    
#     print('dict_tpr_fpr=',dict_tpr_fpr)
    list_reversed_tpr_fpr = [i for i in reversed(list(dict_tpr_fpr.values()))]
#     print('list_reversed_tpr_fpr=', list_reversed_tpr_fpr)
    return list_reversed_tpr_fpr

def calculate_AUC(list_reversed_tpr_fpr):
    # AUC = Area under ROC curve
    AUC = 0
    n_tpr_fpr = len(list_reversed_tpr_fpr)
    for i in range(n_tpr_fpr - 1):        
        if(list_reversed_tpr_fpr[i][1] != list_reversed_tpr_fpr[i + 1][1]):
            # if fpr is changing but tpr is not changing then the area is a square=a*b=(tpr[i+1]-0)*(fpr[i+1]-fpr[i])
            if(list_reversed_tpr_fpr[i][0] == list_reversed_tpr_fpr[i + 1][0]):
                AUC += list_reversed_tpr_fpr[i + 1][0]*(list_reversed_tpr_fpr[i + 1][1] - list_reversed_tpr_fpr[i][1])
            else:
                # if both are changing then the area is a trapezoid=(a+b)*h/2=(tpr[i]+tpr[i+1])*(fpr[i+1]-fpr[i])/2
                AUC += (list_reversed_tpr_fpr[i + 1][0] + list_reversed_tpr_fpr[i][0])*(list_reversed_tpr_fpr[i + 1][1] - list_reversed_tpr_fpr[i][1])/2
        
    return AUC
    
def auc(df, correctlabels):    
    AUC = 0
    class_frequency = dict()
    for i in correctlabels:
        if i not in class_frequency:
            class_frequency[i] = (1/len(correctlabels))
        else:
            class_frequency[i] += (1/len(correctlabels))

    for col in df.columns:
        prediction_vector = df[col]       
        list_reversed_tpr_fpr = get_tpr_fpr(prediction_vector, correctlabels, col)
        area_col = calculate_AUC(list_reversed_tpr_fpr)
#         print('col={}__area_col={}'.format(col, area_col))
        AUC += class_frequency[col] * area_col
    return AUC

#binning

def create_bins(df, nobins, bintype="equal-width"):
    cols = [col for col in df.columns if col != "CLASS" and col != "ID"]
    binning = {}
    if bintype == "equal-width":
#         labels = [i for i in range(nobins)]
        for col in cols:
            df[col], bin_ = pd.cut(df[col], nobins, labels=False, retbins=True)
            # replace the boundaries with inf and -inf to remove NaN values
            bin_[0] = -float("inf")
            bin_[-1] = float("inf")
            binning[col] = bin_
    elif bintype == "equal-size":
#         labels = [i for i in range(nobins-1)]
        for col in cols:
            df[col], bin_ = pd.qcut(df[col], nobins, labels=False, retbins=True,
                                   duplicates="drop", precision=0)
            # replace the boundaries with inf and -inf to remove NaN values
            bin_[0] = -float("inf")
            bin_[-1] = float("inf")
            binning[col] = bin_
    return df, binning

def apply_bins(df, binning):
    cols = [col for col in df.columns if col != "CLASS" and col != "ID"]
    for col in cols:
        df[col] = pd.cut(df[col], binning[col], labels=False)
    return df
```

## 1. Define the class kNN

```python
# Define the class kNN with three functions __init__, fit and predict (after the comments):
#
# Input to __init__: 
# self - the object itself
#
# Output from __init__:
# <nothing>
# 
# This function does not return anything but just initializes the following attributes of the object (self) to None:
# column_filter, imputation, normalization, one_hot, labels, training_labels, training_data, training_time
#
# Input to fit:
# self              - the object itself
# df                - a dataframe (where the column names "CLASS" and "ID" have special meaning)
# normalizationtype - "minmax" (default) or "zscore"
#
# Output from fit:
# <nothing>
#
# The result of applying this function should be:
#
# self.column_filter   - a column filter (see Assignment 1) from df
# self.imputation      - an imputation mapping (see Assignment 1) from df
# self.normalization   - a normalization mapping (see Assignment 1), using normalizationtype from the imputed df
# self.one_hot         - a one-hot mapping (see Assignment 1)
# self.training_labels - a pandas series corresponding to the "CLASS" column, set to be of type "category" 
# self.labels          - a list of the categories (class labels) of the previous series
# self.training_data   - the values (an ndarray) of the transformed dataframe, i.e., after employing imputation, 
#                        normalization, and possibly one-hot encoding, and also after removing the "CLASS" and "ID" columns
#
# Note that the function does not return anything but just assigns values to the attributes of the object.
#
# Input to predict:
# self - the object itself
# df   - a dataframe
# k    - an integer >= 1 (default = 5)
# 
# Output from predict:
# predictions - a dataframe with class labels as column names and the rows corresponding to
#               predictions with estimated class probabilities for each row in df, where the class probabilities
#               are estimated by the relative class frequencies in the set of class labels from the k nearest 
#               (with respect to Euclidean distance) neighbors in training_data
#
# Hint 1: Drop any "CLASS" and "ID" columns first and then apply column filtering, imputation, normalization and one-hot
#
# Hint 2: Get the numerical values (as an ndarray) from the resulting dataframe and iterate over the rows 
#         calling some sub-function, e.g., get_nearest_neighbor_predictions(x_test,k), which for a test row
#         (numerical input feature values) finds the k nearest neighbors and calculate the class probabilities.
#
# Hint 3: This sub-function may first find the distances to all training instances, e.g., pairs consisting of
#         training instance index and distance, and then sort them according to distance, and then (using the indexes
#         of the k closest instances) find the corresponding labels and calculate the relative class frequencies

class kNN():
    
    def _init_(self):
        self.column_filter=None
        self.imputation=None
        self.normalization=None
        self.one_hot=None
        self.labels=None
        self.training_labels=None
        self.training_data=None
        self.training_time=None
    
    def fit(self,df,N_type="minmax" ):
        
        self.now_df=df.copy()
    
        self.training_labels=self.now_df['CLASS']
        self.now_df=self.now_df.drop(columns=["CLASS","ID"]) 
        self.now_df,self.column_filter=create_column_filter(self.now_df)
        self.now_df,self.imputation=create_imputation(self.now_df)
        self.now_df,self.normalization=create_normalization(self.now_df,normalizationtype=N_type)
        self.now_df,self.one_hot=create_one_hot(self.now_df)
        
        
        self.labels=self.training_labels.unique()  
        self.training_data=self.now_df
        
#         print(f"column_filter={self.column_filter}\nimputation={self.imputation}\n")
#         print(f"normalization={self.normalization}\none_hot={self.one_hot}")
#         print(self.labels)
        
    def predict(self,df,k=5):
        self.pred_data=df.copy()
        self.pred_data.drop(columns=["ID","CLASS"],inplace=True)
        #print(self.pred_data)
        
        #self.pred_data=apply_column_filter(self.pred_data, self.column_filter)
        self.pred_data=apply_imputation(self.pred_data, self.imputation)
        self.pred_data=apply_normalization(self.pred_data,self.normalization)
        self.pred_data=apply_one_hot(self.pred_data,self.one_hot)
        
        results=np.empty((len(self.pred_data),len(self.labels)))
        for index,row in self.pred_data.iterrows():
            #print(row)
            results[index]=self.get_nearest_neighbor_predictions(row,k)
            
        predictions=pd.DataFrame(results,columns=self.labels)
        #print(predictions)
        return predictions
    def get_nearest_neighbor_predictions(self,row,k):
        c_df=self.training_data
       
        c_df=c_df-row
        c_df=c_df.apply(lambda x: x**2)
        c_df['distance2']=c_df.apply(lambda x: x.sum(),axis=1)
        c_df['distance']=c_df['distance2']**(1/2)
        c_df.sort_values(by='distance',inplace=True)
        #print(c_df)
        neighbors=c_df.index.tolist()[0:k]
        #print(neighbors)
        probs=np.zeros(len(self.labels))
        
        for n in neighbors:
            for j in range(len(self.labels)):
                if self.training_labels[n]==self.labels[j]:
                    probs[j]+=1
        probs=probs/k
        #print(probs)
        return probs
```

```python
# Test your code (leave this part unchanged, except for if auc is undefined)

glass_train_df = pd.read_csv("glass_train.csv")

glass_test_df = pd.read_csv("glass_test.csv")

knn_model = kNN()

t0 = time.perf_counter()
knn_model.fit(glass_train_df)
print("Training time: {0:.2f} s.".format(time.perf_counter()-t0))

test_labels = glass_test_df["CLASS"]

k_values = [1,3,5,7,9]
#k_values = [3]
results = np.empty((len(k_values),3))

for i in range(len(k_values)):
    t0 = time.perf_counter()
    predictions = knn_model.predict(glass_test_df,k=k_values[i])
    print("Testing time (k={0}): {1:.2f} s.".format(k_values[i],time.perf_counter()-t0))
    results[i] = [accuracy(predictions,test_labels),brier_score(predictions,test_labels),
                  auc(predictions,test_labels)] # Assuming that you have defined auc - remove otherwise

results = pd.DataFrame(results,index=k_values,columns=["Accuracy","Brier score","AUC"])

print()
display("results",results)
```

## 2. Define the class NaiveBayes

```python
# Define the class NaiveBayes with three functions __init__, fit and predict (after the comments):
#
# Input to __init__: 
# self - the object itself
#
# Output from __init__:
# <nothing>
# 
# This function does not return anything but just initializes the following attributes of the object (self) to None:
# column_filter, binning, , class_priors, feature_class_value_counts, feature_class_counts
#
class NaiveBayes():
    
    def _init_(self):
        self.column_filter=None
        self.binning=None
        self.labels=None
        self.class_priors=None
        self.feature_class_value_counts=None
        self.feature_class_counts=None

# Input to fit:
# self    - the object itself
# df      - a dataframe (where the column names "CLASS" and "ID" have special meaning)
# nobins  - no. of bins (default = 10)
# bintype - either "equal-width" (default) or "equal-size" 
#
# Output from fit:
# <nothing>
#
# The result of applying this function should be:
#
# self.column_filter              - a column filter (see Assignment 1) from df
# self.binning                    - a discretization mapping (see Assignment 1) from df
# self.class_priors               - a mapping (dictionary) from the labels (categories) of the "CLASS" column of df,
#                                   to the relative frequencies of the labels
# self.labels                     - a list of the categories (class labels) of the "CLASS" column of df
# self.feature_class_value_counts - a mapping from the feature (column name) to the number of
#                                   training instances with a specific combination of (non-missing, categorical) 
#                                   value for the feature and class label
# self.feature_class_counts       - a mapping from the feature (column name) to the number of
#                                   training instances with a specific class label and some (non-missing, categorical) 
#                                   value for the feature
#
# Note that the function does not return anything but just assigns values to the attributes of the object.


    
    def fit(self,df,nobins=10,bintype="equal-width" ):
        
        self.now_df=df.copy()
        
        self.now_df,self.binning=create_bins(self.now_df, nobins, bintype)
        self.now_df['CLASS']=self.now_df['CLASS'].astype('category')
        self.labels=list(self.now_df["CLASS"].cat.categories)
        #self.now_df=self.now_df.drop(columns=["CLASS","ID"]) 
        
        #self.now_df,self.column_filter=create_column_filter(self.now_df)
        
        self.class_priors=dict(self.now_df['CLASS'].value_counts(normalize = True))
        #print(self.class_priors)
        
        self.feature_class_value_counts={}
        self.feature_class_counts={}
        #print(self.now_df)
        for col in self.now_df.columns:
            if col not in ["CLASS","ID"]:
                temp=self.now_df.dropna(axis=0,how='any',subset=["CLASS",col])
                #print(temp)
                self.feature_class_counts[col]=dict(temp['CLASS'].value_counts())
                group=temp.groupby(["CLASS",col]).size()
                
                self.feature_class_value_counts[col]=dict(group)
                
#         print(self.feature_class_value_counts)
#         print(self.feature_class_counts)

# Input to predict:
# self - the object itself
# df   - a dataframe
# 
# Output from predict:
# predictions - a dataframe with class labels as column names and the rows corresponding to
#               predictions with estimated class probabilities for each row in df, where the class probabilities
#               are estimated by the naive approximation of Bayes rule (see lecture slides)
#
# Hint 1: First apply the column filter and discretization
#
# Hint 2: Iterating over either columns or rows, and for each possible class label, calculate the relative
#         frequency of the observed feature value given the class (using feature_class_value_counts and 
#         feature_class_counts) 
#
# Hint 3: Calculate the non-normalized estimated class probabilities by multiplying the class priors to the
#         product of the relative frequencies
#
# Hint 4: Normalize the probabilities by dividing by the sum of the non-normalized probabilities; in case
#         this sum is zero, then set the probabilities to the class priors

    def predict(self,df):
        self.pred_data=df.copy()
        self.pred_data.drop(columns=["ID","CLASS"],inplace=True)
        self.pred_data=apply_bins(self.pred_data, self.binning)
        
        labels=self.labels
        
        # every class have a corressponding probs for every col(features) in every row(one piece data)
        nlabels,ncol,nrow=len(labels),len(self.pred_data.columns),self.pred_data.shape[0]
        #print(nlabels,ncol,nrow)
        related_freqs=np.zeros([nlabels,nrow,ncol])
        
        for C in range(ncol):
            col=self.pred_data.columns[C]
            for L in range(nlabels):
                label=labels[L]
                    
                for row in range(nrow):
                    value=self.pred_data.iloc[row,C]
                    data=(label,value)
                    #print(data)
                    if (data in self.feature_class_value_counts[col].keys()):
                        feature_value_count=self.feature_class_value_counts[col][data]
                        feature_count=self.feature_class_counts[col][label]

                        # may exist 0 (means no frequency at all )
                        related_freqs[L,row,C]=feature_value_count/feature_count
                    else:
                        related_freqs[L,row,C]=0
        #print(related_freqs)
        
        # product of probs
        related_freqs=related_freqs.prod(axis=2)
        #print(related_freqs)
        
        priors=np.array(list(self.class_priors.values()))
        #print(priors)
        priors=np.tile(priors,nrow).reshape(nrow,nlabels)   # nrow x nlabels
        
        prob=related_freqs.T * priors
        #print(prob.shape)
        #print(prob)
        predictions=pd.DataFrame(prob,columns=self.labels)
        #print( predictions)
        #normalization
        sum_row=prob.sum(axis=1)
#         sum_row[sum_row==0]=1.0
#         result=prob/sum_row.reshape(-1,1)
        
        for i in range(len(prob)):
            for label in self.labels:
                predictions.loc[i,label]=predictions.loc[i,label]/sum_row[i] if sum_row[i]!=0 else predictions.loc[i,label]
        
        #print(result)
        #predictions=pd.DataFrame(result,columns=self.labels)
        
        #print(predictions)
        return predictions
```
